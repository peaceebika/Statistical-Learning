```python
import numpy as np
import pandas as pd
```

```python
# Load Advertising dataset
url = "https://raw.githubusercontent.com/selva86/datasets/master/Advertising.csv"
df = pd.read_csv(url)

# Display first few rows
df.head()
```

```python
# descriptive statistics
df.describe()
```

```python
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Statistics / Regression
import statsmodels.api as sm

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.scatter(df['TV'], df['sales'])
plt.xlabel('TV')
plt.ylabel('sales')
plt.title('sales vs TV')

plt.subplot(1,3,2)
plt.scatter(df['radio'], df['sales'])
plt.xlabel('radio')
plt.ylabel('sales')
plt.title('sales vs radio')

plt.subplot(1,3,3)
plt.scatter(df['newspaper'], df['sales'])
plt.xlabel('newspaper')
plt.ylabel('sales')
plt.title('sales vs newspaper')

plt.tight_layout()
plt.show()
```

# STEP 2

```python
def simple_regression(y, x):
    X = sm.add_constant(x)
    model = sm.OLS(y, X).fit()
    return model
```

```python
# Sale vs TV
model_tv = simple_regression(df['sales'], df['TV'])
model_tv.summary()
```

```python
# Sales vs Radio
model_radio = simple_regression(df['sales'], df['radio'])
model_radio.summary()
```

```python
# Sales vs Newspapper
model_news = simple_regression(df['sales'], df['newspaper'])
model_news.summary()
```

# STEP 3

```python
# Correlation matrix
df[['TV', 'radio', 'newspaper']].corr()
```

```python
# Scatter Matrix
sns.pairplot(df[['TV', 'radio', 'newspaper']])
plt.show()
```

# Step 4 - Multiple Linear Regression Model

```python
# Define predictors and response
X = df[['TV', 'radio', 'newspaper']]
y = df['sales']

# Add intercept
X = sm.add_constant(X)

# Fit multiple linear regression
multi_model = sm.OLS(y, X).fit()

# Model summary
multi_model.summary()
```

# Step 5 - Compare Simple vs Multiple Regression Coefficients

```python
comparison = pd.DataFrame({
    'Simple_TV': model_tv.params,
    'Simple_Radio': model_radio.params,
    'Simple_Newspaper': model_news.params,
    'Multiple': multi_model.params
})

comparison
```

# Step 6

```python
multi_model.pvalues
```

# Step 7 - Final Interpretation

```python
print("R-squared (TV only):", model_tv.rsquared)
print("R-squared (Radio only):", model_radio.rsquared)
print("R-squared (Newspaper only):", model_news.rsquared)
print("R-squared (Multiple):", multi_model.rsquared)
```
