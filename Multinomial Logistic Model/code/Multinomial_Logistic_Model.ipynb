```python
import pandas as pd

dataset_path = 'Wine dataset.csv'
wine_data = pd.read_csv(dataset_path)

wine_data.head()
```

## Dataset Description

The dataset used in my analysis is the **Wine Dataset**, which is commonly used for classification problems.

The dataset contains **178 observations** and **14 variables**. Among these variables, **13 are predictor variables** representing different chemical properties of wine, while **one variable is the target variable called `class`**.

The predictor variables include measurements such as alcohol content, malic acid, ash, magnesium, flavanoids, proanthocyanins, color intensity, hue, and proline. These variables are **continuous numerical variables**.

The target variable **class** is a **categorical variable** with **three classes (1, 2, and 3)** representing three different types of wine.



```python
# dataset shape
wine_data.shape
```

```python
# Checking dataset information
wine_data.info()
```

```python
# Checking if there are missing values
wine_data.isnull().sum()
```

```python
# features and target variable seperated

X = wine_data.drop('class', axis=1)   # predictors
y = wine_data['class']                # target variable

# Check shapes
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split dataset (75% training, 25% testing)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)

# Check shapes
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Multinomial Logistic Regression Model
from sklearn.linear_model import LogisticRegression

# multinomial logistic regression model
model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=1000
)

# model training
model.fit(X_train, y_train)
```

```python
# Check model coefficients
print(model.coef_)
```

```python
import pandas as pd
import numpy as np

# coefficient dataframe
coef_df = pd.DataFrame(
    model.coef_,
    columns=X.columns,
    index=["Class 1", "Class 2", "Class 3"]
)

coef_df
```

## Model Analysis

The multinomial logistic regression model estimates a set of coefficients for each class of wine. Since the dataset contains three wine classes, the model produces three sets of coefficients.

Each coefficient represents the relationship between a predictor variable and the probability of belonging to a specific wine class. A **positive coefficient** indicates that an increase in the corresponding variable increases the likelihood of that class, while a **negative coefficient** decreases the likelihood.

For example, variables such as alcohol content, flavanoids, and color intensity show relatively larger coefficients compared to some other variables. This suggests that these chemical properties have a stronger influence on distinguishing between the wine classes.

Overall, the coefficients help identify which chemical attributes of wine are most important for predicting the wine category.

## Model Validation

The performance of the multinomial logistic regression model was evaluated using the test dataset.

The model achieved an **accuracy of 1.00 (100%)**, indicating that all wine samples in the test set were correctly classified.

The **confusion matrix** shows that the model correctly predicted all observations for each wine class. Specifically, 15 samples of Class 1, 18 samples of Class 2, and 12 samples of Class 3 were correctly classified, with no misclassifications.

The **precision, recall, and F1-score** for all three classes are equal to **1.00**, indicating perfect classification performance. Precision measures how many predicted samples were correct, recall measures how well the model identified all samples of each class, and the F1-score represents the balance between precision and recall.

Overall, the multinomial logistic regression model performed exceptionally well on the wine dataset, suggesting that the chemical features in the dataset are highly effective for distinguishing between the three wine classes.

```python
# predictions on test set
y_pred = model.predict(X_test)

print(y_pred[:10])
```

```python
# Accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Precision, Recall, F1 Score
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
